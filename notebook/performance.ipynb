三种模型的表现评估，要求使用测试集（连续时间片段）

```python
import pandas as pd
df = pd.read_excel("../datasets/flowspeeds.xlsx")
```

```python
import matplotlib.pyplot as plt
import numpy as np
df1 = df["154_FT_101.PV"].copy(True)
df2 = df["154_FT_101.PV"].copy(True)
s1 = df1
s2 = df2.shift(1)
t = df["TIME"]
fig, ax = plt.subplots(figsize=(12, 4))
ax.set_xlabel('Time')
ax.set_ylabel('Flow(m^3/h)')
ax.plot(t, s1,label='Actural')
ax.plot(t, s2, linestyle='--',label='Predict')
plt.legend()
plt.show()
```
